```python
# define the function that computes the sentiment subject
def get_sentiment_subject(text):
    # tokenize and lemmatize the current survey
    stopWords = set(stopwords.words("english"))
    line = re.sub(r'[^\w]', ' ', str(text).lower())
    words = nltk.word_tokenize(line.decode('utf-8'))
    tagged = nltk.pos_tag(words)
    survey = iter([str(word) for word, tag in tagged if tag in 'NN' and len(word) > 1])

    # calculate the sentiment score
    function = 0
    design = 0
    time = 0
    performance = 0

    for word in survey:
        if word in function_list:
            function = function + 1
#             print "function +1", word
        if word in design_list:
            design = design + 1
#             print "design +1", word
        if word in time_list:
            time = time + 1
#             print "time +1", word
        if word in performance_list:
            performance = performance + 1
#             print "performance +1", word

    subjects = ["function", "design", "time", "performance"]
    return max(subjects, key = locals().get)
```
